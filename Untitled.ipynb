```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('masterCSVfmfull.csv')
```

```python
f_yvsn = (data['country'] == 'Poland')
pl_all =data.loc[f_yvsn, ['sex', 'country', 'year', 'generation', 'suicides_no',
       'population', 'suicides/100k pop']]
pl_all['suicides_no'].sum()
```

```python
pl_all
```

```python
f_pl = (pl_all['sex'] == 'female') & (pl_all['suicides_no'] > 0)
pl_f = pl_all.loc[f_pl]
pl_f
```

```python
m_pl = (pl_all['sex'] == 'male') & (pl_all['suicides_no'] > 0)
pl_m = pl_all.loc[m_pl]
pl_m
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10 ,6))


x = pl_m['year']

yM = pl_m['suicides_no']

x_ind = np.arange(len(x))

plt.plot(x_ind, yM, color= 'b', linestyle='--',label='MÄ™Å¼czyÅºni')

xF = pl_f['year']


yF = pl_f['suicides_no']


plt.plot(x_ind, yF, color= 'r', linestyle='--',label='Kobiety')

plt.xticks(x_ind, x)

plt.xlabel('Lata')
plt.ylabel('Liczba samobÃ³jstw')
plt.title('SamobÃ³jstwa: kobiety vs mÄ™Å¼czyÅºni')

plt.legend()

plt.tight_layout()
plt.savefig('MvsK.png')

plt.show()
```
