```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_sb = pd.read_csv('agg_mxminct.csv')
data_sbpkp = pd.read_csv('agg_mxminct_100kpoprst.csv')
```

```python
data_sb = data_sb.drop(['Unnamed: 0'], axis = 1)

data_sbpkp = data_sbpkp.drop(['Unnamed: 0'], axis = 1)
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(12 ,8))

x = data_sb['year']


y1 = data_sb['mean']

plt.subplot(211)
plt.plot(x, y1, color= '#65374D', linestyle='--', marker='o', label='Srednia samobjstw')
plt.xlabel('Lata')
plt.ylabel('Warto redniej')
plt.title('rednia ilo samobjstw na wiecie')
plt.legend()
plt.tight_layout()

y2 = data_sbpkp['mean']

plt.subplot(212)
plt.plot(x, y2, color= '#000034', linestyle='--', marker='o', label='rednia samobjstw na 100 ty. mieszkacw')



plt.xlabel('Lata')
plt.ylabel('Warto redniej')
plt.title('rednia warto stosunku liczby samobjstw na sto tysicy mieszkccw na wiecie')

plt.legend()

plt.tight_layout()
plt.savefig('rednia ilo samobjstw na wiecie.png')

plt.show()
```
